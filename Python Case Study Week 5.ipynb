```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score
```

```python
df = pd.read_csv('C:/Users/user/Desktop/heart.csv')
df
```

```python
standart = StandardScaler()
min_max = MinMaxScaler()
robust = RobustScaler()
```

```python
y = df['target']
x = df.drop('target', axis = 1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 19)
```

```python
x_train = min_max.fit_transform(x_train.to_numpy())
```

```python
x_train
```

```python
x_test = min_max.fit_transform(x_test.to_numpy())
x_test
```

```python
def modelling(model):
    model.fit(x_train,y_train)
    predictions = model.predict(x_test)
    
    acc_scr = accuracy_score(predictions, y_test)
    f1_scr = f1_score(predictions, y_test)
    recall = recall_score(predictions, y_test)
    precision = precision_score(predictions, y_test)
    
    return {"Accuracy Score": acc_scr, 
            "F1 Score": f1_scr, 
            "Recall Score": recall, 
            "Precision Score": precision}
```

```python
modelling(RandomForestClassifier())
```

```python
modelling(SVC())
```

```python
modelling(DecisionTreeClassifier())
```

```python
modelling(LogisticRegression())
```

```python
modelling(KNeighborsClassifier())
```
